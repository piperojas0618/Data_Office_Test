```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('train (1).csv')
df.keys()
```

```python
X = df['Tipo_Trabajo'].values.reshape(len(df), 1)
enc = OneHotEncoder()
a = enc.fit_transform(X).toarray()
a.shape
```

```python
Tipo_Trabajo = []
for i in df.index:
    value = [0] * len(list(df["Tipo_Trabajo"].value_counts().keys()))
        # Se obtiene el indice de la  pocision de la categoria dentro del vector y se pone un uno 
    index = list(df["Tipo_Trabajo"].value_counts().keys()).index(df["Tipo_Trabajo"][i])
    value[index] = 1
    #Se guarda el vector en la lista correspondiente
    Tipo_Trabajo.append(value)
```

```python
Title = []
for value in list(df["Tipo_Trabajo"].value_counts().keys()):
    Title.append("Tipo_Trabajo " + value)
# Se crea un Dataframe con la informacion hotcodding de cada variable
Tipo_Trabajo = pd.DataFrame(Tipo_Trabajo, columns = Title)
```

```python
pd.concat([Tipo_Trabajo, df[['Edad']]], axis=1, join='inner')
```
